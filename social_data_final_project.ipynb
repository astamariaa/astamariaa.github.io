## Cleaning and combining datasets
The main dataset did not have kick off time for each game. We managed to find this data in five seperate datasets for each season. We merged them into one in order to make it easier when we merged it with them main one.

```python
import pandas as pd

seasonal_files = [
    "prem1920.csv",
    "prem2021.csv",
    "prem2122.csv",
    "prem2223.csv",
    "prem2324.csv"
]

# Loading and combining all into one DataFrame
df_list = []

for file in seasonal_files:
    df = pd.read_csv(file)
    df_list.append(df)

# Combining all dataframes into one
df_time = pd.concat(df_list, ignore_index=True)
```

We are selecting only the necessary columns and renaming them to keep consistency with the main dataset.


```python

df_time_slim = df_time[['Date', 'HomeTeam', 'AwayTeam', 'Time']].copy()
df_time_slim.rename(columns={'HomeTeam': 'Home', 'AwayTeam': 'Away'}, inplace=True)
```

Loading the main dataset.

```python
df_prem = pd.read_csv("prem_data.csv")

df_prem['Date'] = pd.to_datetime(df_prem['Date'], errors='coerce')
```

Filtering step needed for cleaning and formatting the dataset.


```python
df_prem_filtered = df_prem[
    (df_prem['Date'] >= '2019-08-01') & 
    (df_prem['Date'] <= '2024-06-30')
].copy()  

df_prem_filtered['Date'] = pd.to_datetime(df_prem_filtered['Date'], errors='coerce')
df_time_slim['Date'] = pd.to_datetime(df_time['Date'], dayfirst=True, errors='coerce')  # UK format
```

Fix known mismatches to get consistency between the two datasets


```python
# Fix known mismatches
team_name_fixes = {
    "Cardiff City": "Cardiff",
    "Leeds United": "Leeds",
    "Leicester City": "Leicester",
    "Luton Town": "Luton",
    "Manchester City": "Man City",
    "Manchester Utd": "Man United",
    "Newcastle Utd": "Newcastle",
    "Norwich City": "Norwich",
    "Nott'ham Forest": "Nott'm Forest",
    "Sheffield Utd": "Sheffield United"
}

df_prem_filtered['Home'] = df_prem_filtered['Home'].replace(team_name_fixes)
df_prem_filtered['Away'] = df_prem_filtered['Away'].replace(team_name_fixes)
df_time_slim['Home'] = df_time_slim['Home'].replace(team_name_fixes)
df_time_slim['Away'] = df_time_slim['Away'].replace(team_name_fixes)

df_prem_filtered = pd.merge(df_prem_filtered, df_time_slim, on=['Date', 'Home', 'Away'], how='left')
```

```python
df_prem_filtered['Home'] = df_prem_filtered['Home'].replace(team_name_fixes).str.strip()

df_prem_filtered['Stadium Capacity'] = df_prem_filtered['Home'].map(venue_capacity)
```

### Impact of key events
In this section we are looking at how the Premier League schedule and attendance was impacted by two major events.

1. Due to Queen Elizabeth II's death in September 2022 some matches were postponed 
2. The FIFA world Cup in December November/December 2022 paused the Premier league for the whole duration of the world cup. 


We group match data by month and year from August to January across all seasons. 

We wanted to see if we could spot any irregularities - like fewer games or attendnace shifts.

References: 

https://www.espn.com/soccer/story/_/id/37631812/premier-league-suspended-death-queen-elizabeth-ii

### Mapping kickoff time slots for each stadium
We are exploring if there are any regional patterns in match scheduling.

```python
# kickoff times
def classify_time(time_str):
    if pd.isna(time_str):
        return 'Unknown'
    hour = int(time_str.split(':')[0])
    if hour < 13:
        return 'Morning'
    elif hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'

df['Kickoff Slot'] = df['Time'].apply(classify_time)

#Find most common slot per stadium
kickoff_popularity = (
    df
    .dropna(subset=['Latitude', 'Longitude', 'Kickoff Slot'])
    .groupby(['Venue', 'Latitude', 'Longitude'])['Kickoff Slot']
    .agg(lambda x: Counter(x).most_common(1)[0][0])
    .reset_index()
)

# Define color mapping
color_map = {
    'Morning': 'orange',
    'Afternoon': 'blue',
    'Evening': 'purple',
    'Unknown': 'gray'
}

# Plot on map
kickoff_map = folium.Map(location=[52.8, -1.5], zoom_start=6.5)

for _, row in kickoff_popularity.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=10,
        color='black',
        fill=True,
        fill_color=color_map[row['Kickoff Slot']],
        fill_opacity=0.8,
        popup=f"{row['Venue']}<br>Most Common: {row['Kickoff Slot']}"
    ).add_to(kickoff_map)

# Add a simple legend
legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; width: 180px; height: 120px; 
background-color: white; border:2px solid grey; z-index:9999; font-size:14px;">
&nbsp;<b>Kickoff Time Slot</b><br>
&nbsp;<span style="color:orange;">●</span> Morning<br>
&nbsp;<span style="color:blue;">●</span> Afternoon<br>
&nbsp;<span style="color:purple;">●</span> Evening<br>
&nbsp;<span style="color:gray;">●</span> Unknown<br>
</div>
"""
kickoff_map.get_root().html.add_child(folium.Element(legend_html))

# Display the map
kickoff_map
```

### 3PM Blackout
The 3PM blackout is a broadcasting rule in the UK that prevents stations from showing the game on live TV at 3PM on Saturdsays. This was implemented to encourage fans to attend games in person rather than watching them from home

In this plot we want to compare the average stadium fill rate for matches that start at 3PM on Saturdays vs all other matches. 
This shows us if the blackout DOES attract more fans.

https://p11.tv/uk-blackout-window-the-history/#:~:text=The%20football%20blackout%20actually%20began,3pm%20matches%20on%20a%20Saturday.

```python

# define hour
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.hour


df['Is 3PM Saturday'] = (
    (df['Day of Week'] == 'Saturday') &
    (df['Hour'] == 15)
)

#3PM vs other
grouped = df.groupby('Is 3PM Saturday')['Stadium Fill %'].mean()

plt.figure(figsize=(7,5))
grouped.plot(kind='bar', color=['orange', 'steelblue'])

plt.xticks([0, 1], ['Other Matches', '3PM Saturday'], rotation=0)
plt.ylabel('Average Stadium Fill %')
plt.title('Does the 3PM Blackout Affect Attendance?')
plt.tight_layout()
plt.show()
```

```python
# Create season column
df['Season'] = df['Date'].apply(lambda x: f"{x.year-1}/{x.year}" if x.month < 7 else f"{x.year}/{x.year+1}")

# Calculate Fill %
df['Fill %'] = (df['Attendance'] / df['Stadium Capacity']) * 100

# Group and pivot for stacked bar
grouped = df.groupby(['Home', 'Season'])['Fill %'].mean().unstack().fillna(0)

# Sort columns chronologically if needed
seasons = sorted(grouped.columns)
grouped = grouped[seasons]

# Plot
fig, ax = plt.subplots(figsize=(14, 7))
bottom = pd.Series([0]*len(grouped), index=grouped.index)

colors = plt.cm.tab20.colors
for i, season in enumerate(seasons):
    ax.bar(grouped.index, grouped[season], bottom=bottom, label=season, color=colors[i % len(colors)])
    bottom += grouped[season]

# Styling
plt.title('Average Stadium Fill % by Team and Season')
plt.xlabel('Team')
plt.ylabel('Fill %')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Season', loc='upper right')
plt.tight_layout()

# Save as image
plt.savefig("myndo.png", dpi=300)
plt.show()
```
